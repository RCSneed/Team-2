```sql
-- Total Disbursements by Apartment SQL Query
SELECT a.SanitizedName, d.DisbursementsId, d.EffectiveOn, d.ExpiresOn, d.TotalAmount, d.num_RecipientAllocations, d.num_Recipients 
FROM SAN_Disbursements d
JOIN SAN_ApartmentNameRef a ON d.ProductsID = a.ProductID
ORDER BY d.ProductsId;
```
