```sql
%%pyspark
df = spark.read.format("csv").option("header","true").load("Files/zip_codes_6.csv")
# df now is a Spark DataFrame containing CSV data from "Files/all_us_zipcodes(in).csv".
display(df)

df.createOrReplaceTempView("zip_codes_6_temp_view")

spark.sql("CREATE TABLE zip_codes_6 AS SELECT * FROM zip_codes_6_temp_view")
```

```sql
WITH combined_table AS(
    SELECT * FROM zip_codes_1
    UNION ALL
    SELECT * FROM zip_codes_2
    UNION ALL
    SELECT * FROM zip_codes_3
    UNION ALL
    SELECT * FROM zip_codes_4
    UNION ALL
    SELECT * FROM zip_codes_5
    UNION ALL
    SELECT * FROM zip_codes_6
)

SELECT SAN_ApartmentNameRef.SanitizedName, SAN_ApartmentNameRef.Zip, combined_table.states_full_name, combined_table.city, combined_table.county
FROM SAN_ApartmentNameRef
JOIN combined_table
on SAN_ApartmentNameRef.Zip = combined_table.code
-- Where SAN_ApartmentNameRef.Zip BETWEEN 37010 AND 38589
ORDER BY combined_table.states_full_name;
```
